```python

import pandas as pd
from metar.Metar import Metar

# Carregar o arquivo com os METARs
df_raw = pd.read_csv("/content/scrm_metars_2015_2025.csv")  # ajuste o caminho se necessário

# Lista para os resultados decodificados
dados = []

for i, row in df_raw.iterrows():
    try:
        metar_text = row['metar'].replace("SCRM", "").strip()
        metar = Metar(metar_text)

        weather_str = None
        if metar.weather:
            weather_str = " ".join([w[0] if isinstance(w, tuple) else str(w) for w in metar.weather])

        dados.append({
            "datetime": pd.to_datetime(row['valid']),
            "temperature_C": metar.temp.value() if metar.temp else None,
            "dewpoint_C": metar.dewpt.value() if metar.dewpt else None,
            "wind_dir_deg": metar.wind_dir.value() if metar.wind_dir else None,
            "wind_speed_kt": metar.wind_speed.value() if metar.wind_speed else None,
            "wind_gust_kt": metar.wind_gust.value() if metar.wind_gust else None,
            "visibility_m": metar.vis.value() if metar.vis else None,
            "pressure_hPa": metar.press.value() if metar.press else None,
            "weather_raw": weather_str if weather_str else None
        })

    except Exception as e:
        print(f"❌ Erro ao decodificar linha {i}: {e}")

# Criar DataFrame final
df_final = pd.DataFrame(dados)

# Salvar o arquivo com os dados decodificados
df_final.to_csv("scrm_metar_decodificado.csv", index=False)
print("✅ Arquivo salvo: scrm_metar_decodificado.csv")

```
