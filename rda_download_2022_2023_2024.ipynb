```python
import xarray as xr
import glob
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import cftime

# 1. Buscar todos os arquivos de 2023 (ordem correta)
files = sorted(glob.glob("CCMP_RT_Wind_Analysis_2023*.nc"))

# 2. Abrir todos os arquivos empilhando ao longo do tempo
print("🔄 Abrindo arquivos...")
ds = xr.open_mfdataset(
    files,
    combine='nested',
    concat_dim='time',
    decode_times=False,  # manter raw para controlar com cftime
    chunks='auto'        # ideal para muitos arquivos grandes
)

# 3. Definir ponto de interesse
latitude = -62.19
longitude = -58.99
ponto = ds.sel(latitude=latitude, longitude=longitude, method='nearest')

# 4. Calcular velocidade e tempo
uwnd = ponto['uwnd'].values
vwnd = ponto['vwnd'].values
velocidade_ms = np.sqrt(uwnd**2 + vwnd**2)
velocidade_knots = velocidade_ms * 1.94384

# 5. Tempo
unidade = ds.time.attrs['units']
tempos_cftime = cftime.num2date(ds.time.values, units=unidade, calendar='standard')
datas = [pd.Timestamp(t.isoformat()) for t in tempos_cftime]

# 6. Criar DataFrame
df = pd.DataFrame({
    'DataHora': datas,
    'Velocidade_knots': velocidade_knots,
    'Uwnd': uwnd,
    'Vwnd': vwnd
})

# 7. Ordenar e salvar
df = df.sort_values(by='DataHora').reset_index(drop=True)
nome_arquivo = f"vento_{latitude:.2f}_{longitude:.2f}_2023.csv"
df.to_csv(nome_arquivo, index=False)
print(f"\n✅ CSV salvo como: {nome_arquivo}")

# 8. Estatísticas
print("\n📊 Estatísticas da Velocidade do Vento (nós):")
print(df['Velocidade_knots'].describe())

# 9. Histograma
plt.figure(figsize=(8, 4))
plt.hist(df['Velocidade_knots'], bins=15, edgecolor='black', alpha=0.7)
plt.title("Histograma da Velocidade do Vento (nós) - 2023")
plt.xlabel("Velocidade do Vento (nós)")
plt.ylabel("Frequência")
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()

# 10. Série Temporal
plt.figure(figsize=(12, 4))
plt.plot(df['DataHora'], df['Velocidade_knots'], color='blue', label='Velocidade (nós)')
plt.xticks(rotation=45)
plt.title(f"Série Temporal - Velocidade do Vento em ({latitude}, {longitude}) - 2023")
plt.xlabel("Data")
plt.ylabel("Velocidade (nós)")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import xarray as xr
import glob
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import cftime

# 1. Buscar todos os arquivos de 2022 a 2024 (ordem correta)
files = sorted(glob.glob("CCMP_RT_Wind_Analysis_2022*.nc") +
               glob.glob("CCMP_RT_Wind_Analysis_2023*.nc") +
               glob.glob("CCMP_RT_Wind_Analysis_2024*.nc"))

if not files:
    raise FileNotFoundError("❌ Nenhum arquivo encontrado para os anos 2022-2024.")

# 2. Abrir todos os arquivos empilhando ao longo do tempo
print(f"🔄 Abrindo {len(files)} arquivos...")
ds = xr.open_mfdataset(
    files,
    combine='nested',
    concat_dim='time',
    decode_times=False,
    chunks='auto'
)

# 3. Definir ponto de interesse
latitude = -62.19
longitude = -58.99
ponto = ds.sel(latitude=latitude, longitude=longitude, method='nearest')

# 4. Calcular velocidade e tempo
uwnd = ponto['uwnd'].values
vwnd = ponto['vwnd'].values
velocidade_ms = np.sqrt(uwnd**2 + vwnd**2)
velocidade_knots = velocidade_ms * 1.94384

# 5. Tempo
unidade = ds.time.attrs['units']
tempos_cftime = cftime.num2date(ds.time.values, units=unidade, calendar='standard')
datas = [pd.Timestamp(t.isoformat()) for t in tempos_cftime]

# 6. Criar DataFrame
df = pd.DataFrame({
    'DataHora': datas,
    'Velocidade_knots': velocidade_knots,
    'Uwnd': uwnd,
    'Vwnd': vwnd
})

# 7. Ordenar e salvar
df = df.sort_values(by='DataHora').reset_index(drop=True)
nome_arquivo = f"vento_{latitude:.2f}_{longitude:.2f}_2022_2024.csv"
df.to_csv(nome_arquivo, index=False)
print(f"\n✅ CSV salvo como: {nome_arquivo}")

# 8. Estatísticas
print("\n📊 Estatísticas da Velocidade do Vento (nós):")
print(df['Velocidade_knots'].describe())

# 9. Histograma
plt.figure(figsize=(8, 4))
plt.hist(df['Velocidade_knots'], bins=20, edgecolor='black', alpha=0.7)
plt.title("Histograma da Velocidade do Vento (nós) - 2022 a 2024")
plt.xlabel("Velocidade do Vento (nós)")
plt.ylabel("Frequência")
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()

# 10. Série Temporal
plt.figure(figsize=(14, 4))
plt.plot(df['DataHora'], df['Velocidade_knots'], color='blue', linewidth=0.8, label='Velocidade (nós)')
plt.xticks(rotation=45)
plt.title(f"Série Temporal - Velocidade do Vento em ({latitude}, {longitude}) - 2022 a 2024")
plt.xlabel("Data")
plt.ylabel("Velocidade (nós)")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
